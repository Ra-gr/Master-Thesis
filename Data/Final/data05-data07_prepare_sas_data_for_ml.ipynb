# Winsorize returns and transform input data ('data06_ranked.parquet')
 - Winsorize returns (to 3*std from mean). This affects roughly 1.44% of our observations.
 - Data transformation: cross-sectional rank the characteristics to the [-1,1] interval.

```python
# Load dataframe
df = pd.read_parquet('data05_preprocessed.parquet')

# Winsorize returns to 3 sigma (+- 3 standard deviations from mean)
df["ret_mean"] = df.groupby("DATE")["ret"].transform("mean")
df["ret_std"] = df.groupby("DATE")["ret"].transform("std")
df["ret_+3s"] = df["ret_mean"] + 3*df["ret_std"]
df["ret_-3s"] = df["ret_mean"] - 3*df["ret_std"]
df.insert(loc=df.columns.get_loc("ret")+1, column="ret_wins", value=df["ret"])
df.loc[ (df["ret_wins"]>df["ret_+3s"]), "ret_wins"] = df["ret_+3s"]
df.loc[ (df["ret_wins"]<df["ret_-3s"]), "ret_wins"] = df["ret_-3s"]
df = df.drop(columns=['ret_mean', 'ret_std', 'ret_+3s', 'ret_-3s'])
df
```

```python
# Save summary statistic
summary = df.describe()
summary = pd.concat([summary, df.reindex(summary.columns, axis=1).agg(['median', 'skew', 'kurt'])], axis=0)
summary.to_excel('summary_statistic.xlsx')
summary
```

```python
## Cross-sectionaly rank the characteristics to the [-1,1] interval

# Exceptions 1: Columns with data that we do not want to change now (includes reprisk rating)
exc_1 = ['DATE', 'public_date', 'datadate', 'fyear', 'rdq', 'gvkey', 'permno', 'reprisk_id', 'sic2', 'FFI49', 'exchcd',
         'ret', 'ret_wins', 'prc', 'vol', 'mve_m', 'reprisk_rating']

# Exceptions 2: Variables with only integer values 0 and 1 -> remain unchanged
exc_2 = ['convind', 'divi', 'divo', 'rd', 'securedind', 'sin']

# List of characteristics for the ranking
cols_trans = list(set(df.columns)-set(exc_1)-set(exc_2))
print(len(cols_trans))

# Transformation of some reprisk characteristics (divide by 100 in order to be in the interval but also to preserve the relations)
df['country_sector_average_01'] = df['country_sector_average']/100
df['current_rri_01'] = df['current_rri']/100
df['peak_rri_01'] = df['peak_rri']/100
df['trend_rri_01'] = df['trend_rri']/100

# Rank transformation
df[cols_trans] = df.groupby('DATE')[cols_trans].rank(pct=True)*2-1

# Save data to parquet
df.to_parquet('data06_ranked.parquet')
df
```

# Create model input data ('data07_model_input.parquet')
 - Calculate excess returns (above risk-free rate)
 - Make sure that the dates are correct (year-month)
 - Drop unused columns
 - Make an industry dummy variable: 68 industries
 - Make a reprisk rating dummy variable: 10 categories [A, AA, AAA, B, BB, BBB, C, CC, CCC, D]

```python
# Load dataframe
df = pd.read_parquet('data06_ranked.parquet')

# Load risk-free rate (1 month t-bill rate from Fama-French)
FF5 = pd.read_csv('F-F_Research_Data_5_Factors_2x3.csv')
FF5['YM_ret'] = pd.to_datetime(FF5['Date'], format='%Y%m').dt.to_period("M")
FF5['rf'] = FF5['RF'] / 100
FF5 = FF5[['YM_ret', 'rf']]
print(FF5.head())

# Merge risk-free rate to corresponding observation (on YM)
df['YM_ret'] = df["DATE"].dt.to_period("M")
df = pd.merge(df, FF5, how='left', on='YM_ret')

# Calculate excess returns
df['ret_ex'] = df['ret_wins'] - df['rf']
df
```

```python
# Make a month_year variable. (Remember: Current 'DATE' variable is the last day in the month of the next month's return)
df["YM"] = (df["DATE"] - pd.DateOffset(months=1)).dt.to_period("M")
df["year"] = (df["DATE"] - pd.DateOffset(months=1)).dt.to_period("Y")

# Keep only relevant columns
info_vars = ['YM', 'year', 'permno', 'gvkey', 'reprisk_id', 'sic2', 'prc', 'vol', 'mve_m']
X_vars = ['absacc', 'acc', 'aeavol', 'age', 'agr', 'baspread', 'beta', 'betasq', 'bm', 'bm_ia', 'cash', 'cashdebt',
          'cashpr', 'cfp', 'cfp_ia', 'chatoia', 'chcsho', 'chempia', 'chinv', 'chmom', 'chpmia', 'chtx', 'cinvest',
          'convind', 'currat', 'depr', 'divi', 'divo', 'dolvol', 'dy', 'ear', 'egr', 'ep', 'gma', 'grcapx', 'grltnoa',
          'herf', 'hire', 'idiovol', 'ill', 'indmom', 'invest', 'lev', 'lgr', 'maxret', 'mom12m', 'mom1m', 'mom36m',
          'mom6m', 'ms', 'mve', 'mve_ia', 'nincr', 'operprof', 'orgcap', 'pchcapx_ia', 'pchcurrat', 'pchdepr',
          'pchgm_pchsale', 'pchquick', 'pchsale_pchinvt', 'pchsale_pchrect', 'pchsale_pchxsga', 'pchsaleinv', 'pctacc',
          'pricedelay', 'ps', 'quick', 'rd', 'rd_mve', 'rd_sale', 'realestate', 'retvol', 'roaq', 'roavol', 'roeq',
          'roic', 'rsup', 'salecash', 'saleinv', 'salerec', 'secured', 'securedind', 'sgr', 'sin', 'sp', 'std_dolvol',
          'std_turn', 'stdacc', 'stdcf', 'tang', 'tb', 'turn', 'zerotrade']
reprisk_vars = ['country_sector_average', 'country_sector_average_01', 'current_rri', 'current_rri_01',
                'peak_rri', 'peak_rri_01', 'trend_rri', 'trend_rri_01', 'reprisk_rating']
Y_vars = ['ret', 'ret_wins', 'ret_ex']
print("Drop the following columns: ", list(set(df.columns).symmetric_difference(set(info_vars+X_vars+reprisk_vars+Y_vars))))
df = df[list(info_vars+X_vars+reprisk_vars+Y_vars)]

# Make reprisk rating & industry dummy variables (from 'sic2')
df = pd.get_dummies(df, columns=["reprisk_rating", "sic2"])

# Save data to parquet
df.to_parquet('data07_model_input.parquet')
df
```
